```python
                                                                # A Simple To-Do List application

import customtkinter
from tkinter import *
from tkinter import messagebox
from tkinter import PhotoImage , Label,Tk
from PIL import Image, ImageTk
import tkinter as tk
from tkinter import ttk

# Initialize the main window
app = tk.Tk()
app.title("To Do List")
app.geometry("800x600")
app.configure(bg="#a99df0")  # Light purple background color

def add_task():
    task = task_entry.get()
    if task:
        tasks_list.insert(0,task)
        task_entry.delete(0, END)
        save_task()
    else:
        messagebox.showerror('Error',"Enter a Task")
def remove_task():
    selected= tasks_list.curselection()
    if selected:
        tasks_list.delete(selected[0])
        save_task()

    else:
        messagebox.showerror('Error',"Choose a Task To Delete")


def save_task():
    with open("Task.txt", "w") as f:
        task = tasks_list.get(0,END)
        for task in task:
            f.write(task + "/n")

def update_task():
    selected = tasks_list.curselection()
    if selected:
        new_task = task_entry.get()
        if new_task:
            tasks_list.delete(selected[0])
            tasks_list.insert(selected[0], new_task)
            task_entry.delete(0, END)
            save_task()
        else:
            messagebox.showerror('Error', "Enter a Task to Update")
    else:
        messagebox.showerror('Error', "Choose a Task to Update")


def load_tasks():
    try:
        with open("task.txt", "r") as f:
            task = f.readlines()
            for task in task:
                tasks_list.insert(0,task.strip)

    except FileNotFoundError:
        #messagebox.showerror('Error',"Cannot load Tasks.")
        pass

# Sidebar frame
sidebar = tk.Frame(app, bg="#f3f0ff", width=200)
sidebar.pack(side="left", fill="y")

# Sidebar content
profile_pic = tk.Label(sidebar, text="Do-it", bg="#f3f0ff", font=("Arial", 16, "bold"))
profile_pic.pack(pady=10)

username = tk.Label(sidebar, text="Hritik Rajshahi", bg="#f3f0ff", font=("Arial", 14))
username.pack()

menu_items = [ ]
for item in menu_items:
    button = tk.Button(sidebar, text=item, bg="#e5dfff", font=("Arial", 12), bd=0)
    button.pack(fill="x", pady=5, padx=5)


# Main content frame
main_content = tk.Frame(app, bg="#d8b5f5", width=600, height=600)
main_content.pack(side="right", fill="both", expand=True)

# Main content heading
focus_label = tk.Label(main_content, text="Today main focus", bg="#d8b5f5", font=("Arial", 16, "bold"))
focus_label.pack(pady=10)

 

#Button 


add_button = customtkinter.CTkButton(app,command=add_task ,text_color='#fff',text='Add Task',fg_color='gray',
                                     hover_color='#06911f',bg_color='#00B2EE' ,width=130)
add_button.place(x=380,y=150)

remove_button = customtkinter.CTkButton(app,command=remove_task, text_color='#fff' ,text='Remove Task',fg_color='#8B0000',
                                     hover_color='#DC143C',bg_color='#00B2EE' )
remove_button.place(x=200,y=550)

update_button = customtkinter.CTkButton(app, command=update_task,   text_color='#fff', text='Update Task', fg_color='blue',
                                        hover_color='#00CED1',  bg_color='#00B2EE' )
update_button.place(x=550, y=550)


# Add task input

task_entry= customtkinter.CTkEntry(app, text_color='#000',fg_color='#fff', border_color='#fff' ,width=280)
task_entry.place(x=310,y=100)

tasks_list = Listbox(app,width=55,height=20 )
tasks_list.place(x=290,y=210)


app.mainloop()
```
